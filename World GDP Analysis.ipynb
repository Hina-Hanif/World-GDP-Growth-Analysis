# DATA CLEANING AND PREPROCESSING

```python
data['Code'] = data['Code'].fillna(data['Code'].mode()[0])
data
```

```python
data = data.rename(columns={
    'Gross domestic product, constant prices - Percent change - Observations': 'GDP_Obs',
    'Gross domestic product, constant prices - Percent change - Forecasts': 'GDP_Forecast',
    '1120034-annotations': 'Annot1',
    '1120032-annotations': 'Annot2'
})
data
```

```python
data['GDP_Obs'] = data['GDP_Obs'].fillna(data['GDP_Obs'].median())
data
```

```python
data = data.drop(columns=['Annot1', 'Annot2', 'GDP_Forecast'])
data
```

# VISUALIZATION

```python
sns.set_theme(style="whitegrid")
plt.rcParams.update({"figure.dpi": 140, "font.size": 11})
PLOT_DIR = Path("plots")
PLOT_DIR.mkdir(exist_ok=True)
```

```python
top_entities = data['Entity'].value_counts().head(6).index.tolist()  
ts = data[data['Entity'].isin(top_entities)].dropna(subset=['Year','GDP_Obs'])
fig, ax = plt.subplots(figsize=(11,5))
palette = sns.color_palette('tab10', n_colors=len(top_entities))
for i, ent in enumerate(top_entities):
    s = ts[ts['Entity'] == ent].sort_values('Year')
    ax.plot(s['Year'], s['GDP_Obs'], marker='o', linewidth=2, label=ent, color=palette[i])
ax.set_title('Time series â€” GDP_Obs for sample entities')
ax.set_xlabel('Year'); ax.set_ylabel('GDP_Obs')
ax.legend(ncol=2)
plt.tight_layout()
plt.show()

fig.savefig(PLOT_DIR / "time_series_sample_entities.png", bbox_inches='tight')
```

```python
highlight = top_entities[0]  
s = data[data['Entity'] == highlight].dropna(subset=['Year','GDP_Obs']).sort_values('Year')
fig, ax = plt.subplots(figsize=(11,4))
ax.plot(s['Year'], s['GDP_Obs'], marker='o', linewidth=2.4, color='#0d47a1')
ax.fill_between(s['Year'], s['GDP_Obs'], alpha=0.12, color='#0d47a1')
if not s.empty:
    mx = s.loc[s['GDP_Obs'].idxmax()]
    last = s.iloc[-1]
    ax.scatter([mx['Year']], [mx['GDP_Obs']], s=120, color='orange', edgecolor='k')
    ax.annotate(f"peak {mx['GDP_Obs']:.2f}\n{int(mx['Year'])}",
                xy=(mx['Year'], mx['GDP_Obs']), xytext=(5,15), textcoords='offset points',
                arrowprops=dict(arrowstyle='->', color='gray', lw=1.1))
    ax.annotate(f"last {last['GDP_Obs']:.2f}\n{int(last['Year'])}",
                xy=(last['Year'], last['GDP_Obs']), xytext=(-70,-30), textcoords='offset points',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
ax.set_title(f"Highlighted trend â€” {highlight}")
ax.set_xlabel('Year'); ax.set_ylabel('GDP_Obs')
plt.tight_layout(); plt.show()
fig.savefig(PLOT_DIR / "highlight_trend.png", bbox_inches='tight')
```

```python
top12 = data['Entity'].value_counts().head(12).index.tolist()
sub = data[data['Entity'].isin(top12)].dropna(subset=['Year','GDP_Obs'])
g = sns.FacetGrid(sub, col='Entity', col_wrap=4, height=2.4, sharey=False)
g.map_dataframe(sns.lineplot, x='Year', y='GDP_Obs', marker='o')
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Year", "GDP_Obs")
plt.tight_layout(); plt.show()
g.fig.savefig(PLOT_DIR / "small_multiples_top12.png", bbox_inches='tight', dpi=160)
```

```python
vals = data['GDP_Obs'].dropna()
fig, ax = plt.subplots(figsize=(9,5))
sns.histplot(vals, bins=60, kde=True, stat='density', edgecolor='white', ax=ax, color='#e76f51', alpha=0.85)
sns.rugplot(vals.sample(min(len(vals), 500)), ax=ax, height=0.03, color='k')  
ax.set_title("Distribution of GDP_Obs â€” Histogram + KDE")
ax.set_xlabel("GDP_Obs"); ax.set_ylabel("Density")
plt.tight_layout(); plt.show()
fig.savefig(PLOT_DIR / "hist_kde_gdp.png", bbox_inches='tight')
```

```python
top6_by_median = (data.groupby('Entity')['GDP_Obs'].median().sort_values(ascending=False).head(6).index.tolist())
sub = data[data['Entity'].isin(top6_by_median)]
fig, ax = plt.subplots(figsize=(9,5))
sns.boxplot(data=sub, x='Entity', y='GDP_Obs', palette='Set2', showfliers=True, ax=ax)
ax.set_title("Boxplot: GDP_Obs distribution â€” top 6 entities by median")
ax.set_xlabel('')
plt.tight_layout(); plt.show()
fig.savefig(PLOT_DIR / "boxplot_top6.png", bbox_inches='tight')
```

```python
sub = data.dropna(subset=['Year','GDP_Obs'])
fig, ax = plt.subplots(figsize=(9,5))
sample = sub.sample(min(len(sub), 3500), random_state=42)
sns.scatterplot(data=sample, x='Year', y='GDP_Obs', alpha=0.45, s=18, ax=ax)
sns.regplot(data=sample, x='Year', y='GDP_Obs', scatter=False, ax=ax, line_kws=dict(color='darkred', linewidth=2))
ax.set_title('Year vs GDP_Obs (sample) â€” scatter + regression')
ax.set_xlabel('Year'); ax.set_ylabel('GDP_Obs')
plt.tight_layout(); plt.show()
fig.savefig(PLOT_DIR / "scatter_reg_year_gdp.png", bbox_inches='tight')
```

```python
data2 = data.dropna(subset=['Year','GDP_Obs']).copy()
data2['Decade'] = (data2['Year'] // 10) * 10
pivot = data2.pivot_table(index='Entity', columns='Decade', values='GDP_Obs', aggfunc='median')
good = pivot.dropna(thresh=1).index 
topN = pivot.loc[good].median(axis=1).sort_values(ascending=False).head(20).index
p = pivot.loc[topN]
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(p, annot=True, fmt=".2f", cmap='rocket', linewidths=0.4, ax=ax)
ax.set_title('Median GDP by Entity (rows) and Decade (cols) â€” top 20 by median')
plt.tight_layout(); plt.show()
fig.savefig(PLOT_DIR / "heatmap_entity_decade.png", bbox_inches='tight')
```
